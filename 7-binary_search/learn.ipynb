```python
def max_min_distance(positions: list[int], m: int) -> int:
    positions.sort()

    def is_valid(space):
        next_space_allowed = positions[0]
        balls_placed = 0

        for pos in positions:
            if pos >= next_space_allowed:
                balls_placed += 1
                next_space_allowed = pos + space

                if balls_placed == m:
                    return True

        return False


    min_space = 1
    max_space = (positions[-1] - positions[0]) // (m - 1) + 1
    print("Max space:", max_space)
    answer = -1

    while min_space <= max_space:
        mid = (min_space + max_space) // 2
        print("mid:", mid)

        if is_valid(mid):
            print("valid")
            answer = mid
            min_space = mid + 1
        else:
            print("not valid")
            max_space = mid - 1

    return answer

            
```

```python
# Example / sanity
assert max_min_distance([1, 2, 8, 12, 17], 3) == 7

# Unsorted input (should sort internally)
assert max_min_distance([8, 1, 17, 12, 2], 3) == 7

# m = 2 -> should spread to ends (max distance = max - min)
assert max_min_distance([0, 5, 9, 13], 2) == 13
assert max_min_distance([1, 100], 2) == 99

# m = len(positions) -> must use all baskets (answer = min adjacent gap)
assert max_min_distance([1, 3, 6, 10], 4) == 2
assert max_min_distance([0, 3, 7, 12, 18], 5) == 3

# Equally spaced
assert max_min_distance([0, 10, 20, 30, 40], 3) == 20
assert max_min_distance([0, 10, 20, 30, 40], 4) == 10

# Clustered positions
assert max_min_distance([1, 2, 3, 1000, 1001, 1002], 3) == 2
assert max_min_distance([1, 2, 3, 1000, 1001, 1002], 4) == 2

# Non-uniform spacing
assert max_min_distance([2, 11, 14, 19, 25], 3) == 11
assert max_min_distance([2, 11, 14, 19, 25], 4) == 6

# Large coordinates
assert max_min_distance([1, 10**9], 2) == 10**9 - 1

print("All tests passed!")
```
