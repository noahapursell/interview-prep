## Problem 1: Remove Duplicates from Sorted Array

**Difficulty**: Easy  
**Category**: Two Pointers (Same Direction, In-Place)

### Problem Statement

Given a **sorted array** `nums`, remove the **duplicates in-place** such that each element appears **only once**, and return the new length.

Do not allocate extra space for another array — you must do this by **modifying the input array in-place** with O(1) extra memory.

### Function Signature

```python
def remove_duplicates(nums: list[int]) -> int:
    pass  # Modify nums in-place and return new length
```

### Example

```
Input: nums = [1, 1, 2]
Output: 2
Modified nums = [1, 2, _]

Input: nums = [0,0,1,1,1,2,2,3,3,4]
Output: 5
Modified nums = [0,1,2,3,4, _, _, _, _, _]
```

The first `k` elements of `nums` should contain the result. It doesn't matter what values are left beyond the first `k`.

### Constraints

- `1 <= len(nums) <= 10⁵`
- `nums` is sorted in **non-decreasing** order


````python
def remove_duplicates(nums: list[int]) -> int:
    slow_pointer = fast_pointer = 1

    while fast_pointer < len(nums):
        if nums[fast_pointer] == nums[slow_pointer - 1]:
            fast_pointer += 1
        else:
            nums[slow_pointer] = nums[fast_pointer]
            fast_pointer += 1
            slow_pointer += 1

    return slow_pointer
````

````python
# Basic cases
nums = [1, 1, 2]
k = remove_duplicates(nums)
assert k == 2
assert nums[:k] == [1, 2]

nums = [0,0,1,1,1,2,2,3,3,4]
k = remove_duplicates(nums)
assert k == 5
assert nums[:k] == [0, 1, 2, 3, 4]

# Already unique
nums = [1, 2, 3, 4, 5]
k = remove_duplicates(nums)
assert k == 5
assert nums[:k] == [1, 2, 3, 4, 5]

# All elements the same
nums = [7, 7, 7, 7, 7]
k = remove_duplicates(nums)
assert k == 1
assert nums[:k] == [7]

# Only one element
nums = [10]
k = remove_duplicates(nums)
assert k == 1
assert nums[:k] == [10]

# Two elements, both different
nums = [1, 2]
k = remove_duplicates(nums)
assert k == 2
assert nums[:k] == [1, 2]

# Large input
nums = [i for i in range(10_000)] * 2
nums.sort()
k = remove_duplicates(nums)
assert k == 10_000
assert nums[:k] == list(range(10_000))
````
